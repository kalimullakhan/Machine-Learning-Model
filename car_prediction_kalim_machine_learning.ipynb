```python
# encoding "Fuel_Type" Column
cardata.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

# encoding "Seller_Type" Column
cardata.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

# encoding "Transmission" Column
cardata.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
# the complete dataset has encoded now
cardata.head()
```

# Splitting the data and Target

```python
# our target variable is "selling_price"
# the "car_name" column doesn't use for my prediction then i will drop
x=cardata.drop(['Car_Name','Selling_Price'],axis=1) 
y=cardata['Selling_Price']
```

# Splitting , Training and Testing the Data

```python
from sklearn.model_selection import train_test_split  

# splitting data arrays into two subsets: for training data and for testing data. 
                                              #With this function, you don't need to divide the dataset manually
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=2) #test_size is 80 is training,20 is testing
```

# Model Training

   # Linear Regression

```python
from sklearn.linear_model import LinearRegression
#LinearRegression fits a linear model with coefficients w = (w1, …, wp) to minimize the residual sum of squares between
                 #the observed targets in the dataset,and the targets predicted by the linear approximation.
```

```python
lin_reg_model= LinearRegression()
lin_reg_model.fit(x_train,y_train) #to fit the model
```

# Model Evaluation

```python
from sklearn import metrics
#The sklearn. metrics module implements several loss, score, and utility functions to measure classification performance. 
        #Some metrics might require probability estimates of the positive class, confidence values, or binary decisions values.
```

```python
training_data_prediction = lin_reg_model.predict(x_train) 
```

```python
training_data_prediction
```

```python
# coffecient of determination regression score function 
```

```python
# R squared Error
error_score=metrics.r2_score(y_train,training_data_prediction)
print(" The error_score  value is")
error_score #error score should be as low as passible
```

## visualiztion the actual price and pridicted price

```python
plt.scatter(y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

## test the model

```python
testing_data_prediction = lin_reg_model.predict(x_test)
```

```python
# R squared Error
error_score=metrics.r2_score(y_test,testing_data_prediction)
print(" The error_score value  is")
error_score #error score should be as low as passible
```

```python
plt.scatter(y_test,testing_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
#This method is usually used in machine learning for the selection of the subset of variables. It provides greater prediction
  #accuracy as compared to other regression models. Lasso Regularization helps to increase model interpretation.
```

```python
# loading the linear regression model
lass_reg_model = Lasso()
```

```python
lass_reg_model.fit(x_train,y_train)
```

# Model Evaluation
